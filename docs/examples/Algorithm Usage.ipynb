# Algorithm Usage

### Setting path and importing modules

```python
import sys
sys.path.append("../../")
import pathlib as pl
from IPython.display import Image 

import antialiasing as aa

images_path = pl.Path("./images")
```

### Importing an image
Image is imported as a numpy array.

```python
image = aa.imp(images_path / "triangle.jpeg")
type(image)
```

### Applying an algorithm
We are applying an uniform grid algorithm to scale the image.

```python
modified=aa.unif_grid_2(image,1.75)
```

### Exporting an image
We will export the modified image as "triange_mod.jpeg".

```python
aa.exp(modified, images_path / "triange_mod.jpeg")
```
